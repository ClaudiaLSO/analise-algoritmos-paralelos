```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('resultados.csv')
df_mean = df.groupby(['Algoritmo', 'Tamanho', 'Modo', 'Threads']).mean().reset_index()

def calculate_speedup(row):
    serial_time = df_mean[(df_mean['Algoritmo'] == row['Algoritmo']) & 
                          (df_mean['Tamanho'] == row['Tamanho']) & 
                          (df_mean['Modo'] == 'Serial')]['TempoMS'].values[0]
    return serial_time / row['TempoMS'] if row['TempoMS'] > 0 else 0

df_parallel = df_mean[df_mean['Modo'] == 'Paralelo'].copy()
df_parallel['Speedup'] = df_parallel.apply(calculate_speedup, axis=1)

# --- GRÁFICO 1: Desempenho O(n²) ---
fig_slow = px.bar(df_mean[df_mean['Algoritmo'].isin(['BubbleSort', 'SelectionSort'])], 
                  x='Algoritmo', y='TempoMS', color='Modo', barmode='group',
                  facet_col='Tamanho', title='Desempenho Algoritmos $O(n^2)$',
                  labels={'TempoMS': 'Tempo Médio (ms)'}, template='plotly_white')
fig_slow.show()

# --- GRÁFICO 2: Escalabilidade O(n log n) ---
fig_fast = px.line(df_parallel[df_parallel['Algoritmo'].isin(['MergeSort', 'QuickSort'])], 
                   x='Threads', y='Speedup', color='Algoritmo', markers=True,
                   facet_col='Tamanho', title='Análise de Speedup (Escalabilidade)',
                   labels={'Speedup': 'Ganho de Desempenho (x)'}, template='plotly_dark')
fig_fast.show()

# --- GRÁFICO 3: Speedup Comparativo (N=10.000) ---
fig_speedup = px.bar(df_parallel[df_parallel['Tamanho'] == 10000], 
                     x='Algoritmo', y='Speedup', color='Threads', barmode='group',
                     title='Análise de Speedup por Algoritmo ($N=10.000$)',
                     labels={'Speedup': 'Ganho de Desempenho (x)'},
                     text_auto='.2f', template='plotly_white')
fig_speedup.add_hline(y=1, line_dash="dash", line_color="red", annotation_text="Limiar de Eficiência ($S=1$)")
fig_speedup.show()

# --- EXIBIÇÃO DOS DADOS ---
print("Amostra dos dados brutos (CSV):")
display(df.head(10))
```
